```python
def orb_sim(img1, img2):
  # SIFT is no longer available in cv2 so using ORB
  orb = cv2.ORB_create()

  # detect keypoints and descriptors
  kp_a, desc_a = orb.detectAndCompute(img1, None)
  kp_b, desc_b = orb.detectAndCompute(img2, None)

  # define the bruteforce matcher object
  bf = cv2.BFMatcher(cv2.NORM_HAMMING, crossCheck=True)
    
  #perform matches. 
  matches = bf.match(desc_a, desc_b)
  #Look for similar regions with distance < 50. Goes from 0 to 100 so pick a number between.
  similar_regions = [i for i in matches if i.distance < 50]  
  if len(matches) == 0:
    return 0
  return len(similar_regions) / len(matches)
```

```python
def structural_sim(img1, img2):

  sim, diff = structural_similarity(img1, img2, full=True)
  return sim
```

```python
def match(img):
  achou = 0
  for f in glob.iglob(r'C:\Users\55619\Desktop\Python Banco de Dados\Crop*.png'):
    image = cv2.imread(f, 0)
    img = np.array(img)
    orb_similarity = orb_sim(img, image)  #1.0 means identical. Lower = not similar
    print("Similarity using ORB is: " + str(orb_similarity) + "\n")
    if orb_similarity > 0.7:
      print('Mesmo item encontrado')
      print(f)
      achou += 1
    else:
      print("Novo Item")
  if achou > 0:
    return True
  else:
    return False
```

```python
import tkinter as tk
import sys
import threading

class Reservabolsas3000:
    def __init__(self, root):
        self.root = root
        self.root.title("Reservabolsas 3000")

        # Configuração do estilo rosa
        self.root.configure(bg="#FFC0CB")  # Definindo a cor rosa como plano de fundo

        # Definindo a geometria da janela
        largura_janela = 400
        altura_janela = 400  # Aumentamos a altura para caber os três botões
        largura_tela = root.winfo_screenwidth()
        altura_tela = root.winfo_screenheight()
        posicao_x = (largura_tela - largura_janela) // 2
        posicao_y = (altura_tela - altura_janela) // 2
        self.root.geometry(f"{largura_janela}x{altura_janela}+{posicao_x}+{posicao_y}")

        # Criando um frame para centralizar os botões
        frame_central = tk.Frame(self.root, bg="#FFC0CB")
        frame_central.pack(expand=True)

        # Variável de controle para indicar se o código está em execução
        self.running = False

        # Definindo o tamanho padrão dos botões
        botao_width = 15
        botao_height = 3

        # Definindo o tamanho da fonte para os botões
        fonte_botao = ("Helvetica", 16, "bold")

        # Criação do botão "Start" estilizado com tamanho padrão e fonte maior
        self.start_button = tk.Button(
            frame_central, text="Start", command=self.start_program, bg="#FFB6C1", fg="white",
            activebackground="#FF69B4", activeforeground="white", font=fonte_botao,
            width=botao_width, height=botao_height
        )
        self.start_button.pack(pady=10)

        # Criação do botão "Stop" estilizado com tamanho padrão e fonte maior
        self.stop_button = tk.Button(
            frame_central, text="Stop", command=self.stop_program, bg="#FFB6C1", fg="white",
            activebackground="#FF69B4", activeforeground="white", font=fonte_botao,
            width=botao_width, height=botao_height
        )
        self.stop_button.pack(pady=10)

        # Criação do botão "Exit" estilizado com tamanho padrão e fonte maior
        self.exit_button = tk.Button(
            frame_central, text="Exit", command=self.exit_program, bg="#FFB6C1", fg="white",
            activebackground="#FF69B4", activeforeground="white", font=fonte_botao,
            width=botao_width, height=botao_height
        )
        self.exit_button.pack(pady=10)

    def start_program(self):
        if not self.running:
            self.running = True
            print("Iniciando código...")
            # Iniciar a execução do código em uma thread separada para não bloquear a interface
            threading.Thread(target=self.program).start()

    def stop_program(self):
        if self.running:
            self.running = False
            print("Parando todos os códigos...")

    def exit_program(self):
        print("Fechando a interface e parando o código...")
        self.root.destroy()
        sys.exit()


if __name__ == "__main__":
    root = tk.Tk()
    app = Reservabolsas3000(root)
    root.mainloop()
```
